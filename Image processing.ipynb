## Horizontal Flip

###### An image can be flipped by calling the flip() function and passing in a method such as FLIP_LEFT_RIGHT for a horizontal flip or FLIP_TOP_BOTTOM for a vertical flip. Other flips are also available

```python
def Horizontal_Flip(self):
    np_image = np.array(io.imread('try.jpg'))
    fig = plt.figure(figsize=(10, 10))
    fig.add_subplot(1, 2, 1)
plt.title('Horizontal Flip')    
plt.imshow(np.flipud(np_image))
   
```

## Convert to Grayscale

The image contains only black, white, and gray colors, in which gray has multiple levels.
   
    Transforms a colour image to a greyscale image by
    taking the mean of the RGB values, weighted
    by the matrix weights
    

```python
def to_grayscale(im, weights = np.c_[0.2989, 0.5870, 0.1140]):
 
    tile = np.tile(weights, reps=(im.shape[0],im.shape[1],1))
    return np.sum(tile * im, axis=2)
```

```python
img = to_grayscale(np_image)


plt.title('Greyscale')
plt.imshow(img, cmap='Greys')
```

## Histogram For Pixel Intensity

#### Lets draw the histogram using a matplotlib hist() function. Check the below code to draw the Pixel Intensity Histogram

```python

img = np.array(io.imread('try.jpg'))
img_flat = img.flatten()
plt.hist(img_flat, bins=200, range=[0, 256])
plt.title("Number of pixels in each intensity value")
plt.xlabel("Intensity")
plt.ylabel("Number of pixels")
plt.show()
```
