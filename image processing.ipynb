# Basic Image Processing using Numpy

## Importing the libraries

```python

from skimage import io
from skimage import color
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import PIL.ImageOps 
```

### For Horizontal Flip and Vertical Flip

#### An image can be flipped by calling the flip() function and passing in a method such as FLIP_LEFT_RIGHT for a horizontal flip or FLIP_TOP_BOTTOM for a vertical flip. Other flips are also available¶

### For Greyscale

#### The image contains only black, white, and gray colors, in which gray has multiple levels.Transforms a colour image to a greyscale image by taking the mean of the RGB values, weighted by the matrix weights

### For Black and White

#### Color images usually use the RGB format which means every pixel is represented by a tuple of three value (red, green and blue) in Python.L mode on the other hand only uses one value between 0-255 for each pixel (8-bit).

## Code for manipulation

```python
k=1
while k==1:

    print('#'*20)
    print("Press 1 to Load the image",end="\n")    #Menu for the manipulation program
    print("Press 2 to Edit the image",end="\n")
    print("Press 3 to Save the image",end="\n")
    print("Press 4 to Quit",end="\n")
    print('#'*20)
    
    pause_time = 0.4

    option=int(input("Enter the your choice"))

    if option==1:                                 #Load image
        np_image=io.imread('try.jpg') 
        np_array=np.array(np_image)
        print("Image is Loaded")


    elif option==2:
        print('#'*20)                  
        print("Press 1 to Horizontal flip",end="\n")
        print("Press 2 to Vertical flip",end="\n")     #Menu for editing the image
        print("Press 3 to Invert colours",end="\n")
        print("Press 4 to Convert to grayscale",end="\n")
        print("Press 5 to Convert to B&W",end="\n")
        print('#'*20)

        edit_option=int(input("Enter the choice"))

        if  edit_option==1:                      #Horizontal Flip
            np_image_hflip=np.fliplr(np_image) 
            plt.title('Horizontal Flip')    
            plt.imshow(np_image_hflip)
            plt.show()

 
        elif  edit_option==2:                    #Vertical Flip
            np_image_vflip=np.flipud(np_image)    
            plt.title('Vertical Flip')
            plt.imshow(np_image_vflip)
            plt.show()

        elif  edit_option==3:                        #Invert Colours
            image = Image.open('try.jpg')
            inverted_image = PIL.ImageOps.invert(image)
            plt.title('Invert Colours')
            plt.imshow(inverted_image)
            plt.show()
            
        elif  edit_option==4:                    #Convert to Grayscale
            img = io.imread('try.jpg')     
            imgGray = color.rgb2gray(img)
            plt.title('Grayscale Image')
            plt.imshow(imgGray)
            plt.show()
            
        elif edit_option==5:                  #Convert to B&W
            img = Image.open('try.jpg')
            bw_img=img.convert("1")
            plt.title('Black and White Image')   
            plt.imshow(bw_img)
            plt.show()
            
        else:
            print("Invalid choice")

    elif option==3:                      #Save the Image
        print("Image is saved")
        plt.imsave('try.jpg',np_array)
        plt.imshow(np_image)        
        plt.show()

    elif option==4:
        print("Quit")
        k=0           #Quit
        break

    else:
        print("Invalid choice")     
    
    
```
